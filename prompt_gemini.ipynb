```python
import os
import google.generativeai as genai
from tabulate import tabulate
from dotenv import load_dotenv

load_dotenv()

# Configure API key
genai.configure(api_key=os.getenv("GEMINI_API_KEY"))
model = genai.GenerativeModel("gemini-2.5-flash")

prompts = [
    "Explain Artificial Intelligence in one sentence.",
    "List three real-world applications of Generative AI.",
    "Describe the difference between AI, ML, and Deep Learning in simple terms.",
    "Write a short motivational quote about learning AI.",
    "What is GenAI."
]

results = []

for i, prompt in enumerate(prompts, start=1):
    response = model.generate_content(prompt)
    results.append([i, prompt, response.text.strip()])

# Display output
# for i, prompt, answer in results:
#     print(f"{i}. {prompt}")
#     print(f"{answer}\n")
print(tabulate(results, headers=["#", "Prompt", "Gemini Output"], tablefmt="grid"))
```
